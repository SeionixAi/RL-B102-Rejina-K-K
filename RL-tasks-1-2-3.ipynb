```python
#recursion
#Writw a recursive function that : *finds the factorial of a number
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
#Docstrings
#Write a function that * uses a docstring
def greet():
    """This function prints a greeting message"""
    print("Hello Python")
greet()
print(greet.__doc__)

```

Functions in Python are objects and Python stores extra metadata using special attributes called dunder attributes (double underscore).
Examples:
__doc__ documentation string
__name__ function name
__defaults__  default arguments
Formal Definition:
__doc__ is a built-in attribute that stores the docstring of a function, class, module, or method.
The docstring is the string written immediately below the function definition

# DATASET 2: Student Performance Dataset

## Purpose: Sorting, value counts, apply, conditional columns


```python
np.random.seed(2)
rows=150
age=np.append(np.random.randint(18,30,int(rows*.8)),[np.nan]*int(rows*.2))
np.random.shuffle(age)
df=pd.DataFrame({
    "Student_id":np.arange(1,rows+1),
    "Age":pd.Series(age,dtype="Int64"),
    "Gender":np.random.choice(["Male","Female","Trans"],rows),
    "Course":np.random.choice(["B.Sc-CS","BCA","BA-English","MCA"],rows),
    "Marks":np.random.randint(35,100,rows)
})
df
```

```python
#sorting data
df.sort_values(by="Marks",ascending=False)
```

```python
#value counts
df["Gender"].value_counts()
```

```python
#apply function
df["Result"]=df["Marks"].apply(lambda x :"Pass" if x>=50 else "fail")
df.head()
```

```python
#grouping
df.groupby("Course")["Marks"].mean()
```

# DATASET 3: Employee Dataset

## Purpose: Rename, drop, mapping, aggregation


```python
import numpy as np
import pandas as pd
np.random.seed(3)
rows=120
df=pd.DataFrame({
    "EmpId":np.arange(1,rows+1),
    "Dept":np.random.choice(["HR","IT","Sales"]),
    "Salary":np.random.randint(30000,150000,rows),
    "Gender":np.random.choice(["Male","female","Trans"],rows),
    "Age":np.random.randint(18,65,rows)
})
df
```

```python
#view data
df.head()
df.tail()
df.sample(5)
```

```python
#inspect structure
df.shape
```

```python
df.columns
```

```python
df.describe
```

```python
df.info()
```

```python
#handling missing values
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df
```

```python
# Use specific placeholders based on data types
df_filled = df.fillna({
    'Age': 25,
    'City': 'Unknown',
    'Salary': df['Salary'].mean()  # Fill using calculation
})
```

```python
# Fill a column using its mean value
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill a column using its median value
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Fill a categorical column using its most frequent value (mode)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
#rename and drop
df.rename(columns={"Dept":"Department"},inplace=True)
df.drop("Age",axis=1,inplace=True)
df
```
