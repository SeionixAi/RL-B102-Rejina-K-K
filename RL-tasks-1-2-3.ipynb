### 2. What is the difference between mutable and immutable data types?

This dictates whether an object's value can be changed after it is created in memory:

* **Mutable:** The object can be modified in place without changing its memory address.
* *Examples:* `list`, `dict`, `set`.


* **Immutable:** The object cannot be changed. Any operation that appears to modify it actually creates a brand-new object in memory.
* *Examples:* `int`, `float`, `string`, `tuple`.


* **ML Perspective:** Passing large mutable objects (like a massive list of data) into functions can lead to accidental modifications downstream. Tuples are often preferred for fixed dimensions (like image shapes `(224, 224, 3)`) to guarantee immutability.

### 6. Difference between `break`, `continue`, and `pass`

### 7. What is recursion? Give a simple example.

Recursion is a programming technique where a function calls itself to solve a smaller sub-problem of the same problem. To avoid infinite loops, it must always contain a **base case** to halt execution.

* **Example (Factorial Calculation):**
```python
def factorial(n):
    if n == 1:  # Base case
        return 1
    return n * factorial(n - 1)  # Recursive call

```


* **ML Perspective:** Recursion is frequently utilized under the hood in decision tree algorithms (like Random Forests) to continuously split nodes into smaller child groups.

### 9. What is the difference between `is` and `==`?

* **`==` (Equality):** Compares the **values** of two objects to see if they are equal.
* **`is` (Identity):** Compares the **memory addresses** of two objects to check if they point to the exact same object in RAM (`id(a) == id(b)`).

```python
list_a = [1, 2, 3]
list_b = [1, 2, 3]

print(list_a == list_b)  # True (Their data contents match)
print(list_a is list_b)  # False (They are stored in two distinct memory blocks)

```

### 1. What is NumPy?

### 2. Why is NumPy faster than Python lists?

NumPy is significantly faster than Python lists due to a concept called **vectorization**. Here is why:

* **Contiguous Memory Allocation:** Python lists are arrays of pointers scattered across your system memory, which causes high latency (cache misses). NumPy arrays are stored in a continuous block of memory, allowing your CPU to fetch data incredibly fast.
* **Homogeneous Data (Single Type):** Python lists can hold different data types, so Python must check the type of *every single element* during an operation. NumPy arrays force all elements to be the same data type (e.g., `float32`), bypassing this overhead.
* **C-Based Backend:** NumPy operations are written in highly optimized C and Fortran, utilizing modern CPU instructions like SIMD (Single Instruction, Multiple Data) to process whole blocks of data simultaneously without Python loop overhead.

---

### 3. What is an ndarray?

An `ndarray` (N-dimensional array) is the core object of the NumPy library. It is a fast, flexible container for large datasets in Python.

An array can have any number of dimensions:

* **1D array:** A vector (e.g., shape `(5,)`)
* **2D array:** A matrix with rows and columns (e.g., shape `(3, 5)`)
* **3D array:** A tensor, often used for color images where you have height, width, and color channels (e.g., shape `(224, 224, 3)`)

### 1. What is vectorization in NumPy?

**Vectorization** is the practice of replacing explicit loops (like `for` and `while`) with batch-level array operations. Instead of processing elements one by one, a vectorized operation instructs the CPU to perform the calculation on the entire array simultaneously.

Under the hood, NumPy achieves this by delegating the operations to highly optimized C code that utilizes **SIMD (Single Instruction, Multiple Data)** instructions at the processor hardware level.

* **Production Context:** If you try to update the weights of a neural network with 1 million parameters using a Python `for` loop, your training will bottleneck and crawl. Vectorization completes that same operation almost instantly.

### 2. What is the difference between `reshape()` and `resize()`?

Both functions alter the dimensions of your data, but they differ fundamentally in how they handle memory and array size:

* **`reshape()`:** Changes the array's dimensions without modifying its actual data or total element count. It returns a *view* of the original data, meaning it doesn't duplicate memory. If the new shape doesn't match the original element count (e.g., trying to turn a 6-element array into a $3 \times 3$ matrix), it will throw an error.
* **`resize()`:** Permanently modifies the original array itself (in-place) and *can* change the total element count. If the new size is larger than the original, it will pad the extra spaces with zeros or repeated data.

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6])

# reshape() creates a new view
matrix = arr.reshape(2, 3) # Success: 2 * 3 = 6 elements

# resize() changes the original array
arr.resize(2, 4) # Original 'arr' becomes a 2x4 matrix padded with zeros

```

### 3. What are NumPy dimensions and axes?

In NumPy, dimensions are referred to as **axes**. The number of axes an array has is called its rank (`ndim`). Understanding axes is critical when applying mathematical operations like summing or averaging across specific directions of a dataset.

* **Axis 0 (Rows):** Runs vertically down the rows. If you compute `np.mean(matrix, axis=0)`, you are collapsing the rows to calculate the mean of each *column*.
* **Axis 1 (Columns):** Runs horizontally across the columns. If you compute `np.mean(matrix, axis=1)`, you are collapsing the columns to calculate the mean of each *row*.
* **ML Example:** If you have a batch of feature data of shape `(32, 10)`—meaning 32 samples and 10 features—taking the mean along `axis=0` gives you the average value of each feature across the entire batch.

Slicing is the syntax used to extract specific sub-regions or chunks of an array without copying the underlying data. It follows the standard notation `[start:stop:step]`.

Because NumPy slices are **views**, any modification you make to a sliced sub-array will directly alter your original dataset.

```python
# Extracting data from a 2D dataset of shape (100, 5)

```

* **ML Production Context:** Slicing is heavily utilized to split datasets into mini-batches during model training, or to separate your independent features ($X$) from your target labels ($y$).

### 1. What is Pandas?

Pandas is a fast, powerful, and flexible open-source data analysis and manipulation library built on top of the Python programming language. ### 2. What is a DataFrame?

A DataFrame is a 2-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

Think of it as an in-memory SQL table or spreadsheet. It consists of three principal components:

* **Data:** The actual values (which can be numbers, strings, booleans, dates, etc.).
* **Columns:** Labels tracking the vertical fields (e.g., `'Age'`, `'Income'`, `'Target'`).
* **Index:** Labels tracking the horizontal rows (by default, integers starting from `0`, but can be customized to timestamps or user IDs).

---

### 3. Difference between Series and DataFrame

The easiest way to understand the difference is through dimensionality:

* **Series:** A 1-dimensional labeled array capable of holding any data type. It represents a single column or a single row of data.
* **DataFrame:** A 2-dimensional labeled data structure. It is essentially a container for multiple `Series` objects that share the exact same row index.

If you select a single column out of a Pandas `DataFrame`, the resulting object is a Pandas `Series`.

---

### 4. What is a CSV?

CSV stands for **Comma-Separated Values**. It is a simple, plain-text file format used to store tabular data. Each line of the file corresponds to a data row, and each value within that row is separated by a comma (or other delimiters like tabs or semicolons).

Because CSV files are completely text-based, lightweight, and human-readable, they serve as the universal medium for sharing datasets across different platforms (SQL databases, Excel, Python scripts). In Pandas, you can load an entire CSV file into a DataFrame instantly using `pd.read_csv('filename.csv')`.

### 2. What is the difference between `loc` and `iloc`?

Both are properties used to slice and extract specific subsets of rows and columns from a DataFrame, but they navigate data using entirely different indexing systems:

* **`loc` (Label-based indexing):** Selects data using explicit row and column text **labels**. When slicing with `loc`, the stop index is **inclusive**.
* **`iloc` (Integer/Position-based indexing):** Selects data using raw, 0-indexed **integer positions** (just like standard Python lists). When slicing with `iloc`, the stop index is **exclusive**.

* **Imputation:** Filling missing slots with replacement values using `.fillna()`. Common algorithmic strategies include substituting missing entries with the column's **mean**, **median**, or **mode**, or propagating values forward/backward (`ffill`/`bfill`).

A statistical plot is a visualization that goes beyond merely mapping coordinates to visually summarizing the underlying **statistical properties, distributions, and relationships** of a dataset.
